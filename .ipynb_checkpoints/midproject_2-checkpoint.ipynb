# Step 4: Feature Engineering and Selection

## 4.1. Feature Engineering
Feature engineering involves creating new input features from your existing data to help your models capture underlying patterns more effectively.

### 4.1.1. Create Time-Based Features
These features can help your model recognize temporal patterns such as trends and seasonality.



```python
df_merged = pd.read_csv('cruelty_free_data/merged_data.csv')
df_merged.head()
```

```python
df_merged['date'] = pd.to_datetime(df_merged['date'])
# Verify the data types
print(df_merged.dtypes)
```

```python
df_merged['month'] = df_merged['date'].dt.month
df_merged['year'] = df_merged['date'].dt.year
```

```python
df_merged['week'] = df_merged['date'].dt.isocalendar().week
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'
df_merged['season'] = df_merged['month'].apply(get_season)
```

```python
df_merged = pd.get_dummies(df_merged, columns=['season'])
```

```python
df_merged['CFC_Lag1'] = df_merged['cruelty_free_cosmetics'].shift(1)
df_merged['CFC_Lag2'] = df_merged['cruelty_free_cosmetics'].shift(2)
df_merged['CFC_Lag3'] = df_merged['cruelty_free_cosmetics'].shift(3)
```

```python
df_merged['Vegan_Cosmetics_Lag1'] = df_merged['vegan_cosmetics'].shift(1)
```

```python
df_merged['CFC_RollingMean_3'] = df_merged['cruelty_free_cosmetics'].rolling(window=3).mean()
df_merged['CFC_RollingMean_6'] = df_merged['cruelty_free_cosmetics'].rolling(window=6).mean()
```

```python
df_merged['CFC_Diff'] = df_merged['cruelty_free_cosmetics'].diff()
```

```python
# Dates during which the campaign was most active
df_merged['Save_Ralph_Campaign'] = df_merged['date'].apply(
    lambda x: 1 if pd.to_datetime('2021-04-06') <= x <= pd.to_datetime('2021-05-31') else 0
)
```

```python
df_merged.dropna(inplace=True)
```

```python
# Exclude 'Date' from the correlation matrix
corr_matrix = df_merged.drop('date', axis=1).corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Updated Correlation Matrix')
plt.show()
```

```python
# Convert to DataFrame
df_corr = pd.DataFrame(corr_matrix)

# Display the DataFrame
print(df_corr)
```

### 4.3.2. Feature Importance Using Random Forest
Use a Random Forest model to determine feature importance.



```python
# Define target variable
y = df_merged['cruelty_free_cosmetics']

# Define features (exclude 'Date' and 'Cruelty_Free_Cosmetics')
feature_columns = df_merged.columns.difference(['date', 'cruelty_free_cosmetics'])
X = df_merged[feature_columns]
```

```python
# Assuming 'Month', 'Year', 'WeekOfYear' are numeric and 'Season_*' are already dummy variables
# If any categorical variables remain, use one-hot encoding or label encoding
```

```python
# Initialize and Train Random Forest
from sklearn.ensemble import RandomForestRegressor

# Initialize the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model
rf_model.fit(X, y)
```

```python
# Get feature importances
importances = rf_model.feature_importances_

# Create a DataFrame
feature_importances = pd.DataFrame({
    'Feature': feature_columns,
    'Importance': importances
})

# Sort by importance
feature_importances.sort_values(by='Importance', ascending=False, inplace=True)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importances)
plt.title('Feature Importances from Random Forest')
plt.show()
```

```python
# Convert to DataFrame
df_fi = pd.DataFrame(feature_importances)

# Display the DataFrame
print(df_fi)
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.impute import SimpleImputer

# Sample data loading (replace this with your actual DataFrame)
# For example:
# X = pd.read_csv('your_data.csv')

# Select numeric features
X_numeric = X.select_dtypes(include=[np.number]).copy()

# Inspect data
print("Data Types:\n", X_numeric.dtypes)
print("\nMissing Values:\n", X_numeric.isnull().sum())
print("\nInfinite Values Present:", np.isinf(X_numeric).values.any())

# Replace infinite values with NaN
X_numeric.replace([np.inf, -np.inf], np.nan, inplace=True)

# Handle missing values
# Option 1: Drop rows with any NaNs
# X_numeric = X_numeric.dropna()

# Option 2: Impute missing values with the mean
imputer = SimpleImputer(strategy='mean')
X_numeric_imputed = pd.DataFrame(imputer.fit_transform(X_numeric), columns=X_numeric.columns)
X_numeric = X_numeric_imputed

# Ensure all data is float
X_numeric = X_numeric.astype(float)

# Verify no missing or infinite values remain
print("\nAfter Cleaning:")
print("Missing Values:\n", X_numeric.isnull().sum())
print("Infinite Values Present:", np.isinf(X_numeric).values.any())
print("Data Types:\n", X_numeric.dtypes)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['Feature'] = X_numeric.columns
vif_data['VIF'] = [
    variance_inflation_factor(X_numeric.values, i) 
    for i in range(len(X_numeric.columns))
]

print("\nVIF Results:")
print(vif_data)
```

Based on feature importance and multicollinearity analysis, select the final set of features.

```python
# Create 'df_final' by selecting the chosen features
df_final = df_merged[selected_features]
df_final
```

```python
# List of features with infinite VIFs
features_to_remove_inf = ['CFC_Diff', 'CFC_Lag1', 'CFC_Lag2', 'CFC_RollingMean_3']

# Drop these features from df_final
df_final = df_final.drop(columns=features_to_remove_inf)

# Update the selected_features list
selected_features = [feature for feature in selected_features if feature not in features_to_remove_inf]
```

```python
df_final.columns
```

```python
max_vif_threshold = 10  # Define your acceptable VIF threshold

while True:
    vif_data = pd.DataFrame()
    vif_data['Feature'] = X_numeric.columns
    vif_data['VIF'] = [
        variance_inflation_factor(X_numeric.values, i) 
        for i in range(len(X_numeric.columns))
    ]
    
    max_vif = vif_data['VIF'].max()
    if max_vif > max_vif_threshold:
        feature_to_remove = vif_data.loc[vif_data['VIF'] == max_vif, 'Feature'].values[0]
        print(f"Removing '{feature_to_remove}' with VIF = {max_vif}")
        X_numeric = X_numeric.drop(columns=[feature_to_remove])
    else:
        print("All VIF values are below the threshold.")
        break

print("\nFinal VIF DataFrame:")
print(vif_data)

```

```python
# Select features based on previous VIF analysis
selected_features = ['CFC_Lag3', 'Save_Ralph_Campaign', 'cruelty_free_makeup', 'elf', 'week']

```

```python
X_final = X[selected_features]
len(X_final)
```

### 5.3.1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Initialize the model
lr_model = LinearRegression()

# Train the model
lr_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred_lr = lr_model.predict(X_test)
```

```python
# Calculate evaluation metrics
mse_lr = mean_squared_error(y_test, y_pred_lr)
mae_lr = mean_absolute_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

# Print the results
print("Linear Regression Performance:")
print(f"Mean Squared Error (MSE): {mse_lr:.2f}")
print(f"Mean Absolute Error (MAE): {mae_lr:.2f}")
print(f"R-squared (R): {r2_lr:.2f}")
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Initialize the model
dt_model = DecisionTreeRegressor(random_state=42)

# Train the model
dt_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred_dt = dt_model.predict(X_test)
```

```python
# Calculate evaluation metrics
mse_dt = mean_squared_error(y_test, y_pred_dt)
mae_dt = mean_absolute_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

# Print the results
print("\nDecision Tree Regressor Performance:")
print(f"Mean Squared Error (MSE): {mse_dt:.2f}")
print(f"Mean Absolute Error (MAE): {mae_dt:.2f}")
print(f"R-squared (R): {r2_dt:.2f}")
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Initialize the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)
```

```python
# Calculate evaluation metrics
mse_rf = mean_squared_error(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Print the results
print("\nRandom Forest Regressor Performance:")
print(f"Mean Squared Error (MSE): {mse_rf:.2f}")
print(f"Mean Absolute Error (MAE): {mae_rf:.2f}")
print(f"R-squared (R): {r2_rf:.2f}")
```

```python
# Create a DataFrame to compare models
performance_df = pd.DataFrame({
    'Model': ['Linear Regression', 'Decision Tree', 'Random Forest', 'XGBoost'],
    'MSE': [mse_lr, mse_dt, mse_rf, mse_xgb],
    'MAE': [mae_lr, mae_dt, mae_rf, mae_xgb],
    'R_squared': [r2_lr, r2_dt, r2_rf, r2_xgb]
})

# Round the metrics for readability
performance_df['MSE'] = performance_df['MSE'].round(2)
performance_df['MAE'] = performance_df['MAE'].round(2)
performance_df['R_squared'] = performance_df['R_squared'].round(2)

print("\nModel Performance Comparison:")
print(performance_df)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Initialize the Random Forest model
rf = RandomForestRegressor(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    scoring='neg_mean_squared_error'
)
```

```python
# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
print("\nBest Hyperparameters for Random Forest:")
print(grid_search.best_params_)
```

```python
# Use the best estimator
best_rf_model = grid_search.best_estimator_

# Make predictions
y_pred_best_rf = best_rf_model.predict(X_test)

# Evaluate the model
mse_best_rf = mean_squared_error(y_test, y_pred_best_rf)
mae_best_rf = mean_absolute_error(y_test, y_pred_best_rf)
r2_best_rf = r2_score(y_test, y_pred_best_rf)

# Print the results
print("\nTuned Random Forest Regressor Performance:")
print(f"Mean Squared Error (MSE): {mse_best_rf:.2f}")
print(f"Mean Absolute Error (MAE): {mae_best_rf:.2f}")
print(f"R-squared (R): {r2_best_rf:.2f}")
```

```python
# Create a new DataFrame for the new row
new_row = pd.DataFrame({
    'Model': ['Tuned Random Forest'],
    'MSE': [round(mse_best_rf, 2)],
    'MAE': [round(mae_best_rf, 2)],
    'R_squared': [round(r2_best_rf, 2)]
})

# Concatenate the new row to the existing DataFrame
performance_df = pd.concat([performance_df, new_row], ignore_index=True)

# Display the updated DataFrame
print("\nUpdated Model Performance Comparison:")
print(performance_df)
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression

# Define features and target
X = df_final[['cruelty_free_makeup', 'cruelty_free_skincare', 'vegan_cosmetics', 
        'cosmetics', 'vegan', 'the_body_shop', 'lush', 'elf']]
y = df_final['cruelty_free_cosmetics']

# Initialize the model
lr = LinearRegression()

# Define k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation
cv_scores = cross_val_score(lr, X, y, cv=kf, scoring='r2')

print(f"Cross-Validation R Scores: {cv_scores}")
print(f"Mean R: {cv_scores.mean():.2f}  {cv_scores.std():.2f}")

#df_merged
#Cross-Validation R Scores: [0.77620491 0.81281631 0.97402508 0.77735803 0.81822515]
#Mean R: 0.83  0.07

#df_final
```

```python
from sklearn.model_selection import cross_validate, KFold

# Define the k-fold cross-validation strategy
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Define scoring metrics
scoring = {
    'r2': 'r2',
    'mse': 'neg_mean_squared_error',  # scikit-learn returns negative MSE
    'mae': 'neg_mean_absolute_error'  # scikit-learn returns negative MAE
}

# Perform cross-validation
cv_results = cross_validate(
    estimator=lr_model,
    X=X_train,
    y=y_train,
    cv=kfold,
    scoring=scoring,
    return_train_score=False
)

# Extract and convert negative scores to positive
r2_scores = cv_results['test_r2']
mse_scores = -cv_results['test_mse']  # Convert to positive
mae_scores = -cv_results['test_mae']  # Convert to positive

# Calculate mean and standard deviation
r2_mean = r2_scores.mean()
r2_std = r2_scores.std()

mse_mean = mse_scores.mean()
mse_std = mse_scores.std()

mae_mean = mae_scores.mean()
mae_std = mae_scores.std()

# Create a summary DataFrame
summary_df = pd.DataFrame({
    'Metric': ['R Score', 'Mean Squared Error (MSE)', 'Mean Absolute Error (MAE)'],
    'Mean': [f"{r2_mean:.4f}", f"{mse_mean:.2f}", f"{mae_mean:.2f}"],
    'Std Dev': [f"{r2_std:.4f}", f"{mse_std:.2f}", f"{mae_std:.2f}"]
})

print("\nCross-Validation Results:")
print(summary_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="whitegrid")

# Plot R Scores
plt.figure(figsize=(10, 4))
sns.barplot(x=[f"Fold {i+1}" for i in range(len(r2_scores))], y=r2_scores, palette="Blues_d")
plt.title('R Scores Across Folds')
plt.ylabel('R Score')
plt.ylim(0, 1)
plt.show()

# Plot MSE Scores
plt.figure(figsize=(10, 4))
sns.barplot(x=[f"Fold {i+1}" for i in range(len(mse_scores))], y=mse_scores, palette="Greens_d")
plt.title('MSE Across Folds')
plt.ylabel('Mean Squared Error')
plt.show()

# Plot MAE Scores
plt.figure(figsize=(10, 4))
sns.barplot(x=[f"Fold {i+1}" for i in range(len(mae_scores))], y=mae_scores, palette="Oranges_d")
plt.title('MAE Across Folds')
plt.ylabel('Mean Absolute Error')
plt.show()
```
